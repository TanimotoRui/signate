```python
import texthero as hero
import pandas as pd
import re
```

```python
train = pd.read_csv('input/train.csv')
test = pd.read_csv('input/test.csv')
#sub = pd.read_csv('input/sample_submit.csv')
```

# texthero„Åßclean

```python
def remove_html(text):
    html_pattern = re.compile('<.*?>')
    return html_pattern.sub(r'', text)
```

```python
for i in range(len(train)):
  train.loc[i, 'cleaned_text'] = remove_html(train.loc[i, 'html_content'])

for i in range(len(test)):
  test.loc[i, 'cleaned_text'] = remove_html(test.loc[i, 'html_content'])
```

```python
train['cleaned_text'] = hero.clean(train['cleaned_text'])
test['cleaned_text'] = hero.clean(test['cleaned_text'])
```

```python
train.to_csv('07_cleaned_train.csv')
test.to_csv('07_cleaned_test.csv')
```
